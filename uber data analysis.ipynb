```python
#Raushan kumar
#importing all the neccessary tools for visualisation and analysis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
#input data file

data=pd.read_csv("other-American_B01362.csv")
data.head()
```

```python
data=data.unstack().unstack()
```

```python
data=data.dropna()
```

```python
data
```

```python
data=data.unstack().unstack()
```

```python
data
```

```python
data['DATE']=pd.to_datetime(data['DATE'])
data['dayofweeks']=data['DATE'].dt.dayofweek
data['weekdayname']=data['DATE'].dt.weekday_name
day=data.pivot_table(index=['weekdayname'],values='dayofweeks',aggfunc='count')
dayavg=day/12
dayavg

```

```python
dayavg.plot()
```

```python
dayavg.plot(kind='bar')
plt.ylabel('Average Rides Per Day')
plt.title('Average Rides per Day vs Day of Week')
```
